```python
import pandas as pd
import requests
import re
import json
pd.set_option("display.max_columns",None)
import re
```

```python
with open('proyecto_json_generos/GeneroAccion.json', 'r') as fichero:
    GeneroAccion= json.load(fichero)
```

```python
GeneroAccion
```

```python
accion = []
```

```python
for action in GeneroAccion:
    tupla_accion = tuple(action)
    accion.append(tupla_accion)
```

```python
accion
```

```python
df = pd.DataFrame(accion)

df.head()
```

```python
df.to_csv("accion.csv")
```
