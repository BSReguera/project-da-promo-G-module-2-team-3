EXTRACCION DATOS - TABLA OSCARS 

Extracción datos Wikipedia

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = "https://es.wikipedia.org/wiki/Premios_%C3%93scar"
res_oscars = requests.get(url)

sopa = BeautifulSoup(res_oscars.content, "html.parser")

tablas = sopa.find_all("table")
```

Extracción valores en tabla

```python
tabla = tablas[2]
```

```python
filas = tabla.findAll("tr")[1:]
```

```python
resultados = []

for fila in filas:
    
    div_fila = fila.text.split("\n\n")

    if len(div_fila) == 10:
        resultados.append(list(div_fila[1:-4]))
    elif len(div_fila) == 8:
        resultados.append(list(div_fila[1:-2]))
    elif len(div_fila) == 9:
        resultados.append(list(div_fila[1:-3]))
```

```python
resultados_2000 = resultados[71:-1]
```

Generación DF e introducción de columnas id

```python
df = pd.DataFrame(resultados_2000)
```
