```python
import numpy as np
import json
import pandas as pd
```

```python
df_drama = pd.read_csv('Archivos/actores_drama.csv')
df_accion = pd.read_csv('Archivos/actores_accion.csv')
df_comedia = pd.read_csv('Archivos/actores_comedia.csv')
```

```python
actores = pd.concat([df_drama, df_accion, df_comedia], axis=0, ignore_index=True)
```

```python
actores
```

```python
Total_actores = actores[['tconst','nconst']]
```

```python
Total_actores
```

```python
Total_actores.drop_duplicates()
```

```python
Total_actores2 = Total_actores.rename(columns={'tconst':'id_film','nconst':'id_actor'})
```

```python
Total_actores2
```

```python
Total_actores2.to_csv('Actores.csv', index=False)
```
