```python
# Importaciones necesarias
import pandas as pd
import random
from bs4 import BeautifulSoup
import requests
import json
import csv
from selenium import webdriver  
from webdriver_manager.chrome import ChromeDriverManager  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.support.ui import Select 
from time import sleep
from selenium.webdriver.common.by import By

# Configuracion, establece una opción de Pandas para mostrar todas las columnas de un DataFrame.
pd.set_option('display.max_columns', None)
```

```python
# Solicitud de apertura del archivo
with open ('../Fase1_datos/lista_accion.json', 'r') as fichero:
    lista_accion = json.load (fichero)
```

```python
# Observamos la longitud de la lista accion
len(lista_accion)
```

```python
# Creamos una variable llamada busqueda_actores que haga referencia a la lista accion para la extraccion de los datos
busqueda_actores = lista_accion[0:100]
```

```python
# EJECUTAR SOLO LA PRIMERA VEZ
# Creaccion de listas vacias para posteriormente completarlas, apendearlas con la informacion necesaria
lista_tuplas1 = []
lista_errores = []
```

```python
#Paso1 
# Inicializar el navegador y maximizar la página
driver = webdriver.Chrome()
driver.get('https://www.google.com/')
driver.maximize_window()

# Uso de sleep dinámico
sleep(random.randint(1,3))

# Aceptar las cookies si es necesario
driver.find_element('css selector', '#W0wltc > div').click()

# Uso de sleep dinámico
sleep(random.randint(1,3))



#Paso 2
# Iterar a través de las peliculas, para la extraccion de los actores
for pelicula in busqueda_actores:
     # Escribe en el buscador
    driver.find_element('css selector', '#APjFqb').send_keys(pelicula[2],' ', pelicula[3],' ','reparto',Keys.ENTER)
    try:
        # Uso de sleep dinámico
        sleep(random.randint(1,3))
        
        # Uso de sleep dinámico
        sleep(random.randint(1,3))
        lista_actores = [pelicula[0],[]]
        for actor in range(1,11):
            info_actores = []
            info_actores.append(driver.find_element('css selector', f'#rcnt > div.M8OgIe > div > div > div > div > div > div.uciohe > div:nth-child({actor})').text.split('\n'))
            for dato in info_actores:
                lista_actores[1].append(dato[0])
        lista_tuplas1.append(tuple(lista_actores))
        driver.find_element('css selector', '#APjFqb').clear()
    except:
        lista_errores.append(pelicula[0])
        driver.find_element('css selector', '#APjFqb').clear()
        continue
```
